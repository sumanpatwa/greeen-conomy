# CO2 PRODUCTIVITY

```python
### CO2 PRODUCTIVITY  Included are CO2 emissions from combustion of  coal, oil, natural gas and other fuels.
###  GDP generated per unit of CO2 emitted (USD/kg)
### Demand-based emissions reflect the CO2 from energy use emitted during the various stages of production of goods and services consumed in domestic final demand, irrespective of where the stages of production occurred
i = 0
x = 'start'
c = ['b','g','r','y','v']
plt.subplots(figsize=(16,6)) 
L =list()
while True:
    if x != 'stop':
        x = input("Enter Country  ")
        k = b[b.COU == x]
        j = k[k.VAR == 'CO2_DBPROD']
        t = k[k.VAR == 'CO2_PBPROD']
        plotdata = pd.DataFrame({'Year':j['Year'],'Value':j['Value']})
        plotdata1 = pd.DataFrame({'Year':t['Year'],'Value':t['Value']})
        plt.plot(plotdata['Year'].astype(int),plotdata['Value'],label = x,color = c[i])
        plt.plot(plotdata1['Year'].astype(int),plotdata1['Value'],label = x,color = c[i],linestyle = 'dashdot')
        L.append(x+" Demand")
        L.append(x+" Production")
        i=i+1
    else:
        L.pop()
        L.pop()
        plt.legend(L)        
        break
```

# Energy productivity.

```python
## Energy productivity.
##Energy productivity is calculated as GDP per unit of TPES (USD/toe). It
#reflects, at least partly, efforts to improve energy efficiency and to reduce
#carbon and other atmospheric emissions. Together with energy intensity, these
#indicators also reflect structural and climatic factors. 
## Energy intensity, TPES per capita
i = 0
x = 'start'
c = ['b','g','r','y','b']
plt.subplots(figsize=(16,6)) 
L =list()
while True:
    if x != 'stop':
        x = input("Enter Country  ")
       
        k = b[b.COU == x]
        t = k[k.VAR == 'NRG_PROD']
        plotdata1 = pd.DataFrame({'Year':t['Year'],'Value':t['Value']})
        
        plt.plot(plotdata1['Year'].astype(int),plotdata1['Value'],label = x,color = c[i],linestyle = 'dashdot')
        L.append(x+" Productivity")
        i=i+1
    else:
        L.pop()
        plt.legend(L)        
        break
# ENERGY PRODUCTIVITY is a measure of the economic benefit we receive from each unit of energy we use. It is calculated by dividing TOTAL ECONOMIC OUTPUT (e.g., GDP or revenue) by the AMOUNT OF ENERGY
#CONSUMED (e.g., barrels of oil equivalent, or kilowatt
#hours of electricity)
```
